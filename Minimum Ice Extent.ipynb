```python
data = {}
septseries = []
septlist = []
yearlist = []
fig, axs = plt.subplots(1,2)
for year in range(1978, 2017):
    dseries = [] #data series
    for month in range(0,12):
        dseries.append(df[str(year)][month])
    data[str(year)] = pd.Series(dseries,index = range(1,13)) 
    ts = pd.DataFrame(data, index = range(1,13))
    septseries.append(float(ts[str(year)][9]))
    septlist.append(9)
    yearlist.append(year)
    yearsept = pd.Series(data = septseries, index = yearlist)
    ser = pd.Series(data = septseries, index = septlist).to_frame()
    ser.reset_index(inplace = True)
    ser.columns = ["Month","Million Square Km"]
    ax = ser.plot.scatter(x = "Month", y = "Million Square Km", s  =  75, c = "b")
    ts.plot(figsize = (15,15), ax = ax)
    plt.xlabel("Month")
    plt.ylabel("Million Square Km")
    plt.title("Minimum Sea Ice Extent")
    plt.savefig("MSIE" + str(year))
    #plt.show()
    plt.clf()
    yearsept.plot(figsize = (30,15), xticks = range(1978,2017), yticks = range(0,7),  title = "Minimum Sea Ice In September")
    plt.xlabel("Year")
    plt.ylabel("Million Square Km")
    plt.savefig("MSIESeptember" + str(year))  
    #plt.show()
    plt.clf()
    
    
    
    
```
