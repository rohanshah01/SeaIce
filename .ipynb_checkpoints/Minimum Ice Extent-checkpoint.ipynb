```python
data = {}
for year in range(1978, 2017):
    dseries = [] #data series
    for month in range(0,12):
        dseries.append(df[str(year)][month])
    data[str(year)] = pd.Series(dseries,index = ["Jan","Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov","Dec"]) 
    ts = pd.DataFrame(data, index = ["Jan","Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov","Dec"])
    ts.plot(figsize = (20,20))
    plt.xlabel("Month")
    plt.ylabel("Million Square Km")
    plt.title("Minimum Sea Ice Extent")
    plt.savefig("MSIE" + str(year))
```
