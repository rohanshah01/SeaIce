```python
data = []
for year in range(1979, 2016):
    for month in range(0,12):
        data.append(df[str(year)][month])
ts = pd.Series(data, index = pd.date_range("1/1/1979", periods = ((2016-1979)*12), freq = "M"))
ts.plot(figsize = (30, 5))
plt.xlabel("Year")
plt.ylabel("Penis")
plt.title("Minimum Sea Ice Extent")
plt.show()
```
